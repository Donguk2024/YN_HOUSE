```python
import os
import shutil

# 원본 폴더 경로
folder_path = 'C:/Users/chg20/OneDrive/Desktop/YN_HOUSE/raw_data/아이디어탱크/ternel_type_house'

# 새 폴더 경로
new_folder_path = 'C:/Users/chg20/OneDrive/Desktop/YN_HOUSE/source_data/ternel_type_warehouse'

# 새 폴더가 존재하지 않는다면 생성
if not os.path.exists(new_folder_path):
    os.makedirs(new_folder_path)

# 폴더 내의 파일 목록을 가져옴
files = os.listdir(folder_path)

identification = []
# 파일을 순회하면서 작업 수행
for idx, filename in enumerate(files):
    # 파일 이름을 띄어쓰기를 기준으로 분할
    parts = filename.split()
    # 마지막 부분만 추출
    new_filename = f"{idx+1}.hwp"
    # 원본 파일 경로
    original_file_path = os.path.join(folder_path, filename)
    # 새 파일 경로
    new_file_path = os.path.join(new_folder_path, new_filename)
    # 파일 복사
    shutil.copy(original_file_path, new_file_path)
    identification.append([parts[-1], idx])

    print(f"Copied '{original_file_path}' to '{new_file_path}'")
```

```python
import win32com.client as win32
import win32gui
import os

def setup_hwp():
    """한글 애플리케이션을 설정하고 창을 숨김"""
    hwp = win32.gencache.EnsureDispatch("HWPFrame.HwpObject")
    hwp.RegisterModule("FilePathCheckDLL", "SecurityModule")
    hwnd = win32gui.FindWindow(None, '한/글')  # 창 제목에 따라 조정이 필요할 수 있음
    if hwnd:
        win32gui.ShowWindow(hwnd, 0)  # 창 숨기기
    return hwp

def convert_single_hwp_to_pdf(hwp, hwp_path, pdf_path):
    """단일 HWP 파일을 PDF로 변환"""
    hwp.Open(hwp_path)  # 파일 열기
    hwp.SaveAs(pdf_path, "PDF")  # PDF로 저장

def convert_folder_hwp_to_pdf(source_folder, target_folder):
    """지정된 폴더의 모든 HWP 파일을 PDF로 변환"""
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)
    
    hwp = setup_hwp()
    files = os.listdir(source_folder)

    for file in files:
        if file.endswith('.hwp'):
            file_path = os.path.join(source_folder, file)
            pdf_path = os.path.join(target_folder, file.replace('.hwp', '.pdf'))
            convert_single_hwp_to_pdf(hwp, file_path, pdf_path)

    hwp.Quit()  # 한컴오피스 종료

# 폴더 경로 지정
first_source_data_folder = 'C:/Users/Kdw/Desktop/nodejs/yn_house/hwp_source'
second_source_data_folder = 'C:/Users/Kdw/Desktop/nodejs/yn_house/pdf'

# 변환 함수 실행
convert_folder_hwp_to_pdf(first_source_data_folder, second_source_data_folder)
```

```python
import PyPDF2
import re
import os

def extract_text_from_pdf(pdf_path):
    """PDF 파일로부터 텍스트를 추출하는 함수"""
    with open(pdf_path, 'rb') as file:  # PDF 파일을 바이너리 읽기 모드로 열기
        reader = PyPDF2.PdfReader(file)
        text = []

        for page in reader.pages:  # 모든 페이지에 대해 반복
            text.append(page.extract_text())  # 페이지의 텍스트를 추출하고 리스트에 추가

    return "\n".join(filter(None, text))  # None값 제거 후 모든 텍스트를 하나의 문자열로 결합

def remove_space_before_parenthesis(text, keyword = '식'):
    # ')' 뒤의 글자가 한글일 때 공백 제거
    text = re.sub(r'\)\s+([가-힣])', r')\1', text)
    
    # ')' 앞의 글자가 한글일 때 공백 제거
    text = re.sub(r'([가-힣])\s+\)', r'\1)', text)

    # '(' 앞의 글자가 한글일 때 공백 제거
    text = re.sub(r'([가-힣])\s+\(', r'\1(', text)

    # '(' 뒤의 글자가 한글일 때 공백 제거
    text = re.sub(r'\(\s+([가-힣])', r'(\1', text)

    # 한글과 한글 사이의 공백 제거 (단, '식' 포함 단어는 제외)
    pattern = r'([가-힣])(?!{})(\s+)(?!{})([가-힣])'.format(keyword, keyword)
    text = re.sub(pattern, r'\1\3', text)

    # 한글 뒤에 숫자가 오는 경우 공백 추가
    text = re.sub(r'([가-힣])(\d)', r'\1 \2', text)
    return text

# 사용 예시

for i in range(1, 777):
    pdf_path = f'C:/Users/Kdw/Desktop/nodejs/yn_house/pdf/{i}.pdf'

    if not os.path.exists(pdf_path):
        print("File {pdf_path} does not exist.")
        continue

    text = extract_text_from_pdf(pdf_path)
    extracted_text = remove_space_before_parenthesis(text)
    print(extracted_text)
```

```python
import pandas as pd

def find_first_occurrence(text, substrings):
    for substring in substrings:
        try:
            return text.index(substring)
        except ValueError:
            continue
    return -1

substrings = ['품    명', '품명']

product_list_start_index = find_first_occurrence(extracted_text,substrings)  # '품명'의 시작 인덱스를 찾음
cleaned_text = extracted_text[product_list_start_index:]  # '품명' 이후의 텍스트를 추출

# 데이터를 줄 단위로 분리
lines = cleaned_text.strip().split('\n')

# 각 줄을 공백을 기준으로 분리하여 DataFrame에 추가
df = pd.DataFrame([line.split(maxsplit=4) for line in lines], columns=['품명', '규격', '수량', '단가', '금액'])
df = df.drop(0)

df
```

```python
import pandas as pd
import re

idx = df[(df['품명'] == '시공비') | (df['품명'] == '인건비') | (df['품명'] == '인건비평')].index[0] - 1

df_first_three_rows = df.loc[:idx].copy()

def split_spec(spec):
    if spec is None or spec.strip() == '':
        return [' '] * 5
    
    parts = spec.split('x')
    if len(parts) < 5:
        parts += [' '] * (5 - len(parts))  # 부족한 차원을 '0'으로 채움
    return parts

df_first_three_rows[['규격_길이', '규격_지름', '규격_높이','규격_두께','규격_원주']] = df_first_three_rows['규격'].apply(split_spec).tolist()

df_first_three_rows = df_first_three_rows.drop(columns=['규격'])

df_first_three_rows
```

```python
#열에서 @뒤의 숫자 제거
df_first_three_rows['규격_높이'] = df_first_three_rows['규격_높이'].apply(lambda x: re.sub('@.*', '', x))

# 열에서 영어 제거
df_first_three_rows['규격_길이'] = df_first_three_rows['규격_길이'].apply(lambda x: re.sub('[^0-9.]', '', x))
df_first_three_rows['규격_지름'] = df_first_three_rows['규격_지름'].apply(lambda x: re.sub('[^0-9.]', '', x))
df_first_three_rows['규격_높이'] = df_first_three_rows['규격_높이'].apply(lambda x: re.sub('[^0-9.]', '', x))
df_first_three_rows['규격_두께'] = df_first_three_rows['규격_두께'].apply(lambda x: re.sub('[^0-9.]', '', x))

#수량, 단가, 금액에서 , 제거
df_first_three_rows['수량'] = df_first_three_rows['수량'].apply(lambda x: re.sub(',', '', x) if x is not None else x)
df_first_three_rows['단가'] = df_first_three_rows['단가'].apply(lambda x: re.sub(',', '', x) if x is not None else x)
df_first_three_rows['금액'] = df_first_three_rows['금액'].apply(lambda x: re.sub(',', '', x) if x is not None else x)

df_first_three_rows
```

```python
# 한글 추출 및 이동 함수 정의
def move_korean_from_spec_to_name(row):
    # 규격_길이에서 한글 부분 추출
    match = re.search(r'[가-힣]+', row['규격_길이'])
    if match:
        korean_text = match.group(0)
        # 품명에 한글 부분 추가
        row['품명'] = korean_text + ' ' + row['품명']
        # 규격_길이에서 한글 부분 제거
        row['규격_길이'] = re.sub(r'[가-힣]+', '', row['규격_길이']).strip()
    return row

# DataFrame에 함수 적용
df_first_three_rows = df_first_three_rows.apply(move_korean_from_spec_to_name, axis=1)

df_first_three_rows
```

```python
def update_values(row):  
    if pd.isna(row['금액']):
        quantity_str = row['수량']
        unit_price = int(row['단가'])
        first_digit = int(quantity_str[0])
        remaining_digits = int(quantity_str[1:])

        if first_digit * remaining_digits == unit_price:
            row['금액'] = str(unit_price)
            row['단가'] = str(remaining_digits)
            row['수량'] = str(first_digit)

        try:
            unit_length = int(row['규격_길이'])
            if unit_length * int(quantity_str) == unit_price:
                row['금액'] = str(unit_price)
                row['단가'] = str(quantity_str)
                row['수량'] = str(unit_length)
                row['규격_길이'] = " "
        except (ValueError, TypeError):
            pass  # '규격_길이' 값이 float로 변환 불가능하거나 null인 경우 예외 처리 

    return row

df_first_three_rows = df_first_three_rows.apply(update_values, axis = 1)

df_first_three_rows.to_csv('1_terner_warehouse.csv', index=False, encoding='utf-8-sig')  # CSV 파일로 저장

df_first_three_rows
```
